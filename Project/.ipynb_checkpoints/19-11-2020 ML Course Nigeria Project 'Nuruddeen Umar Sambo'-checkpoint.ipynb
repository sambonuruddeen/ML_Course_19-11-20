# Preprocessing

- Are there any duplicated values?
- Do we need to do feature scaling?
- Do we need to generate new features?
- Split Train and Test dataset. (0.7/0.3)

```python
#Find duplicates
dup = data.duplicated(['volatile acidity', 'citric acid', 'chlorides', 'density'], keep='last' )
data[dup][:10]
```

```python
#drop duplicates
data.drop_duplicates(['volatile acidity', 'citric acid', 'chlorides', 'density'], keep='last' )
```

```python
#separate Label from features
y_new = data['quality']
X_new = data.drop('quality', axis = 1)
X_new.head()
```

```python
#scaling the data
from sklearn.preprocessing import StandardScaler
X_scaled = StandardScaler().fit_transform(X_new)
X_scaled
```

# ML Application

- Define models.
- Fit models.
- Evaluate models for both train and test dataset.
- Generate Confusion Matrix and scores of Accuracy, Recall, Precision and F1-Score.
- Analyse occurrence of overfitting and underfitting. If there is any of them, try to overcome it within a different section.

```python
from sklearn.model_selection import train_test_split
X_trainS, X_testS, y_trainS, y_testS = train_test_split(X_scaled,y_new, test_size=0.3, random_state=123)
```

```python
print(len(X_testS))
print(len(y_testS))
```

```python
#define & fit model
from sklearn.linear_model import LinearRegression
modelScaled = LinearRegression(normalize=False)
modelScaled.fit(X_trainS, y_trainS)
print("Score of the train set",modelScaled.score(X_trainS,y_trainS))
print("Score of the test set",modelScaled.score(X_testS,y_testS))
```

```python
#Evaluate
def adj_r2 (X,y,model):
    r_squared = model.score(X,y)
    return(1 - (1-r_squared)*(len(y)-1)/(len(y)-X.shape[1]-1))

print("Adj. R2 of the train set",adj_r2(X_trainS,y_trainS,modelScaled))
print("Adj. R2 of the test set",adj_r2(X_testS,y_testS,modelScaled))
```

```python
importance = modelScaled.coef_
for i in range(len(importance)):
    print("Feature", data.columns[i], "Score:", importance[i])
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import  cross_validate
from sklearn.metrics import classification_report, confusion_matrix

#Scaling and outlier removed
X_train, X_test, y_train, y_test = train_test_split(X_scaled,y_new, test_size=0.3, random_state=42)

models = LogisticRegression(random_state=42, n_jobs=-1)
cv = cross_validate(models,X_train,y_train,cv = 8, n_jobs=-1, return_estimator=True)

print("Mean training accuracy: {}".format(np.mean(cv['test_score'])))
print("Test accuracy: {}".format(cv["estimator"][0].score(X_test,y_test)))
```

```python
pred = cv["estimator"][0].predict(X_testS)

cm = confusion_matrix(y_testS, pred)
plt.figure(figsize=(12, 8))
ax =sns.heatmap(cm, square=True, annot=True, cbar=False)
ax.xaxis.set_ticklabels(["False","True"], fontsize = 12)
ax.yaxis.set_ticklabels(["False","True"], fontsize = 12, rotation=0)
ax.set_xlabel('Predicted Labels',fontsize = 15)
ax.set_ylabel('True Labels',fontsize = 15)
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier(max_depth=6, random_state=50)
clf.fit(X_trainS,y_trainS)
print("Accuracy of train:",clf.score(X_trainS,y_trainS))
print("Accuracy of test:",clf.score(X_testS,y_testS))
```

```python
from sklearn.metrics import precision_score, recall_score, accuracy_score, classification_report, f1_score
pred = clf.predict(X_testS)
print(classification_report(y_testS,pred))
```

```python
print("Precision = {}".format(precision_score(y_test, pred, average='macro')))
print("Recall = {}".format(recall_score(y_test, pred, average='macro')))
print("Accuracy = {}".format(accuracy_score(y_test, pred)))
print("F1 Score = {}".format(f1_score(y_test, pred,average='macro')))
```
